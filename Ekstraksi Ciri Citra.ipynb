```
import cv2
import math
import numpy as np
from matplotlib import pyplot as plt 
from skimage.feature import greycomatrix, greycoprops
from PIL import Image
from PIL import GifImagePlugin
```

```
def ciriOrdeSatu(image):
  img = cv2.imread(image)
  img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
  His = cv2.calcHist([img],[0],None,[256],[0,256])
  His = His/His.sum()
  His=His.flatten()

  I = [i for i in range(256)]
  ciriMean = His*I
  ciriMean = ciriMean.sum()
  # CiriENT = 

  ciriVar = 0
  for i in I:
    ciriVar = ciriVar + (((i-ciriMean)**2)*His[i])

  CiriSKEW = 0
  for i in I:
    CiriSKEW = CiriSKEW +(((i-ciriMean)**3)*His[i])
  CiriSKEW = (1/(ciriVar**(3/2)))*CiriSKEW

  ciriENT = 0
  for i in range(256):
    if His[i]!=0:
      t = math.log(His[i],2)
    else:
      t=0;
    ciriENT = ciriENT + (His[i]*t)
  ciriENT = ciriENT * -1

  print('Fitur Mean    : ',ciriMean)
  print('Fitur Variance: ',ciriVar)
  print('Fitur Skewness: ',CiriSKEW)
  print('Fitur Entropy: ',ciriENT)

  plt.plot(His)
  plt.show()
```
